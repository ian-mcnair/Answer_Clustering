# Visualizations

```python
# Plotting accuracy scores across every dataset, for each method of automated grading
scores = pd.read_csv(
    'comparisons.csv'
)

plt.figure(figsize = (25,8))
sns.scatterplot(
    data = scores.sort_values(by = 'score'),
    x = 'dataset',
    y = 'score',
    hue = 'type'
)
sns.lineplot(
    x = scores.dataset,
    y = scores[scores.type == 'Automated'].score.mean(),
    
)

sns.lineplot(
    x = scores.dataset,
    y = scores[scores.type=='Clustering'].score.mean(),
    
)

sns.lineplot(
    x = scores.dataset,
    y = scores[scores.type=='Classification'].score.mean(),
    
)
plt.xticks(rotation = 90)
plt.ylim(0,1)
plt.title('Distribution of Accuracy Scores')
plt.show()
```

```python
# Mean Accuracy
print(f'''Automated: {scores[scores.type == 'Automated'].score.mean()}''')
print(f'''Clustering: {scores[scores.type=='Clustering'].score.mean()}''')
print(f'''Classification: {scores[scores.type=='Classification'].score.mean()}''')
```

```python
# Box and Sqarmplot
plt.figure(figsize = (10,7))
sns.boxplot(
    data = scores,
    x = 'type',
    y = 'score',
    palette = 'Set2'
)
sns.swarmplot(
    data = scores,
    x = 'type',
    y = 'score',
    color = "0.25"
)
plt.ylim(0,1)
plt.title("Methodology Comparison")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.show()
```
