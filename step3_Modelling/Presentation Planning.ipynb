# Visualizations

```python
# Plotting accuracy scores across every dataset, for each method of automated grading
scores = pd.read_csv(
    'acc_comparisons.csv'
)

plt.figure(figsize = (25,8))
sns.scatterplot(
    data = scores.sort_values(by = 'score'),
    x = 'dataset',
    y = 'score',
    hue = 'type',
    hue_order = ['Clustering','Classification', 'Combination' ]
)
sns.lineplot(
    x = scores.dataset,
    y = scores[scores.type == 'Clustering'].score.mean(),
    
)

sns.lineplot(
    x = scores.dataset,
    y = scores[scores.type=='Classification'].score.mean(),
    
)

sns.lineplot(
    x = scores.dataset,
    y = scores[scores.type=='Combination'].score.mean(),
    
)
plt.xticks(rotation = 90)
plt.ylim(0,1)
plt.ylabel("Accuracy", fontsize = 18)
plt.xlabel("Dataset", fontsize = 18)
plt.title('Score Distribution', fontsize = 22)
plt.show()
# Mean Accuracy
print(f'''Clustering: {scores[scores.type=='Clustering'].score.mean()}''')
print(f'''Classification: {scores[scores.type=='Classification'].score.mean()}''')
print(f'''Combination: {scores[scores.type == 'Combination'].score.mean()}''')


print("MEDIAN")
print(f'''Clustering: {scores[scores.type=='Clustering'].score.median()}''')
print(f'''Classification: {scores[scores.type=='Classification'].score.median()}''')
print(f'''Combination: {scores[scores.type == 'Combination'].score.median()}''')


# Box and Sqarmplot
plt.figure(figsize = (10,7))
sns.boxplot(
    data = scores,
    x = 'type',
    y = 'score',
    palette = 'Set2',
    order = ['Clustering','Classification', 'Combination' ]
)
sns.swarmplot(
    data = scores,
    x = 'type',
    y = 'score',
    color = "0.25",
    order = ['Clustering','Classification', 'Combination' ]
)
plt.ylim(0,1)
plt.title("Methodology Comparison")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.show()
```
